## üîπ ¬øPor qu√© es fundamental la estad√≠stica en Machine Learning?

La estad√≠stica y la probabilidad son pilares **cr√≠ticos** de la Inteligencia Artificial (IA). En **IA** y **Machine Learning (ML)** trabajamos con **vastas cantidades de datos** (a menudo denominados *big data*):
*   Los **likes** en redes sociales.
*   Los **precios** de las casas en una ciudad.
*   Las **notas** de los alumnos de una clase.
*   Las **palabras** m√°s usadas en un texto.

Antes de usar estos datos en un modelo inteligente, necesitamos **entenderlos y describirlos bien**. La estad√≠stica act√∫a como la **primera herramienta matem√°tica** para el procesamiento de datos:

*   **Resumir y Describir:** Necesitamos conocer las propiedades estad√≠sticas del conjunto de datos: ¬øcu√°les son los **valores m√°s comunes** (media)? ¬øC√≥mo se dispersan los datos? Medidas como la **varianza y la desviaci√≥n t√≠pica** son esenciales para comprender la distribuci√≥n de los datos y cuantificar la incertidumbre.
*   **Identificar Anomal√≠as:** ¬øHay datos extra√±os o muy diferentes al resto (*outliers*)? La estad√≠stica nos permite identificar estos puntos at√≠picos, t√≠picamente aquellos que se desv√≠an m√°s de dos **desviaciones t√≠picas** de la media.
*   **Relaciones y Generalizaci√≥n:** ¬øDos variables est√°n relacionadas (ejemplo: m√°s horas de estudio ‚Üí mejores notas)? El *Machine Learning* busca modelos que puedan **generalizar** a partir de estos patrones aprendidos para predecir resultados en datos nuevos, un proceso fundamentalmente respaldado por la **inferencia estad√≠stica**.
*   **Preparaci√≥n de Datos:** Se utilizan t√©cnicas estad√≠sticas de **normalizaci√≥n** o **estandarizaci√≥n** para ajustar los datos a una media de cero y una desviaci√≥n t√≠pica de uno. Esto asegura que todas las caracter√≠sticas tengan una escala similar y que ninguna domine el proceso de aprendizaje debido a su magnitud.

La estad√≠stica es clave porque nos ayuda a **resumir, describir y explorar** la informaci√≥n antes de entrenar un modelo, transformando los datos brutos en conocimiento √∫til.

# üé≤ Cap√≠tulo 2: Probabilidad b√°sica: El Lenguaje de la Incertidumbre

La **probabilidad** es el **marco matem√°tico que trata de la verosimilitud (likelihood) de que ocurran los eventos** en un experimento o sistema aleatorio. Nos proporciona las herramientas para **cuantificar la incertidumbre**.

Se expresa como un n√∫mero real que siempre se encuentra entre **0 y 1**:

*   **0** significa que el evento es **imposible** y nunca suceder√° (ej. sacar un 8 en un dado de 6 caras).
*   **1** significa que el evento es **cierto** o seguro, y siempre ocurrir√°.
*   Valores intermedios ($0 < P(E) < 1$) indican que el evento es **incierto**. Por ejemplo: $0.5$ (o 50%) indica igual probabilidad de ocurrencia que de no ocurrencia.

---

## Ejemplos de Eventos y Probabilidad en ML

La probabilidad es fundamental en la Inteligencia Artificial (IA) ya que los sistemas a menudo tienen que operar bajo **incertidumbre o informaci√≥n incompleta**:

1.  **Moneda (Clasificaci√≥n Binaria):**
    *   Cara: 50% (0.5)
    *   Cruz: 50% (0.5)
    *   *Concepto ML:* Un problema de **clasificaci√≥n binaria** (como spam o no spam) puede modelarse con la **Distribuci√≥n de Bernoulli**, que describe un suceso con solo dos posibles resultados.

2.  **Dado de 6 caras (Variables Discretas):**
    *   Probabilidad de sacar un 3: 1 de 6 $\rightarrow \frac{1}{6} \approx 0.166$.
    *   Probabilidad de sacar un n√∫mero par: 3 de 6 $\rightarrow \frac{3}{6} = 0.5$.
    *   *Concepto ML:* El resultado de un lanzamiento de dado es un ejemplo de **variable aleatoria discreta**, que toma un n√∫mero finito o contable de valores.

3.  **Clima (Frecuencia relativa):**
    Si en Le√≥n suele llover 100 d√≠as al a√±o, la probabilidad de que un d√≠a cualquiera llueva es:
    $$\frac{100}{365} \approx 0.27 \rightarrow 27\%$$

---

## C√≥mo calcular probabilidades simples (Interpretaci√≥n Frecuentista)

El c√°lculo de probabilidades simples que asume que todos los resultados son **igualmente probables** se conoce como la **Definici√≥n Cl√°sica o de Laplace**:

$$
P = \frac{\text{Casos favorables}}{\text{Casos posibles}}
$$

Ejemplo:
En una baraja espa√±ola (40 cartas), la probabilidad de sacar un **as** es:
$$P(\text{As}) = \frac{4}{40} = 0.1 \rightarrow 10\%$$

### üìå El Enfoque Frecuentista y la Ley de los Grandes N√∫meros

Esta definici√≥n cl√°sica se relaciona con la **interpretaci√≥n frecuentista de la probabilidad**: la probabilidad de un evento se describe como su **frecuencia relativa** entre un n√∫mero infinito de intentos.

El **valor l√≠mite** de esta frecuencia relativa, a medida que el n√∫mero de intentos tiende a infinito, es la verdadera probabilidad subyacente. Este principio es formalizado por la **Ley de los Grandes N√∫meros**.

$$
P(\text{evento}) = \lim_{n \to \infty} \frac{\text{n√∫mero de ocurrencias}}{\text{n√∫mero de intentos}}
$$

### Ejemplo en Python: Probabilidad Te√≥rica vs. Experimental

Vamos a mostrar c√≥mo la **probabilidad te√≥rica** (0.5) se acerca a la **probabilidad experimental** (lo que obtenemos al simular), ilustrando el concepto de la **Ley de los Grandes N√∫meros**.





`````python
import random

# Lanzamos una moneda 10 veces
resultados = [random.choice(["cara", "cruz"]) for _ in range(10)]
print("Resultados:", resultados)

# Probabilidad experimental de cara
prob_cara = resultados.count("cara") / len(resultados)
print("Probabilidad aproximada de cara:", prob_cara)
`````

`````python
import numpy as np
from scipy.stats import bernoulli
import matplotlib.pyplot as plt

# Simulamos 1000 lanzamientos de una moneda (Bernoulli con p=0.5)
n_tosses = 1000
# 0 para cruz, 1 para cara
coin_tosses = [bernoulli.rvs(p=0.5) for _ in range(n_tosses)]

# Calculamos la frecuencia relativa (promedio) acumulada en cada paso
averages = [np.mean(coin_tosses[:k+1]) for k in range(n_tosses)]

# Representaci√≥n
with plt.style.context("seaborn-v0_8"):
    plt.figure(figsize=(10, 5))
    plt.title("Frecuencia relativa de lanzamientos de moneda")
    plt.xlabel("N√∫mero de lanzamientos")
    plt.ylabel("Frecuencia relativa")

    # Plotea los promedios (frecuencia experimental)
    plt.plot(range(n_tosses), averages, linewidth=3)

    # Plotea la probabilidad te√≥rica (0.5)
    plt.plot([-100, n_tosses+100], [0.5, 0.5], c="k")
    plt.xlim(-10, n_tosses+10)
    plt.ylim(0, 1)
    plt.show()
`````

# üìä 3 Estad√≠stica descriptiva

La **estad√≠stica descriptiva** constituye la primera herramienta matem√°tica utilizada por los sistemas de IA. Su funci√≥n es **resumir y entender los datos**, permitiendo a los modelos transformar datos brutos en percepciones significativas.

## 1. Medidas de tendencia central: Media, Mediana y Moda

Cuando tenemos un **grupo de n√∫meros**, hay tres valores que suelen interesarnos mucho:  
- **Media (Mean)** ‚Üí el promedio.  
- **Mediana (Median)** ‚Üí el valor del medio.  
- **Moda (Mode)** ‚Üí el valor que m√°s se repite.  

Estos valores nos ayudan a **resumir la informaci√≥n** de un conjunto de datos sin tener que mirar cada n√∫mero por separado.

---

## Ejemplo: velocidades de coches
Hemos registrado la velocidad (km/h) de 13 coches:

```
[99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86]

```

---

### üîπ La Media (Mean) o Valor Esperado ($\bar{x}$ o $\mu$)

La media es el promedio aritm√©tico y es la medida de tendencia central m√°s fundamental. En el contexto de la probabilidad y el Aprendizaje Autom√°tico, la media corresponde al **valor esperado** ($\mathbb{E}[X]$), que representa el valor promedio de la variable aleatoria a largo plazo.

La f√≥rmula de la media para una poblaci√≥n finita de $n$ datos, $x_1, x_2, \ldots, x_n$, se define como:

$$\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i$$

En nuestro caso:

$
\text{Media} = \frac{99+86+87+88+111+86+103+87+94+78+77+85+86}{13} = 89.77
$


**Relevancia en IA (C√°lculo Multivariable):**

*   La media es un valor central utilizado para resumir datos y establecer una "l√≠nea base".
*   Las desviaciones respecto a la media se utilizan para calcular la **Varianza** y la **Desviaci√≥n Est√°ndar**, que son medidas cruciales para cuantificar la incertidumbre en las predicciones de los modelos. Por ejemplo, el **Error Cuadr√°tico Medio (MSE)**, una funci√≥n de p√©rdida esencial en la regresi√≥n lineal, mide la varianza de los errores de predicci√≥n.


En Python podemos calcularla con `numpy.mean()`.



---

### üîπ Mediana (Median)
La Mediana es el valor central de un conjunto de datos ordenados [Query]. A diferencia de la Media, la Mediana no se ve afectada por los **valores at√≠picos (*outliers*)**, que son valores significativamente distintos al resto.
```

77, 78, 85, 86, 86, 86, 87, 87, 88, 94, 99, 103, 111

```

Aqu√≠, la mediana es **87**, porque est√° en el centro.  
üëâ Importante: ¬°los datos **deben estar ordenados** primero!

Si hay un n√∫mero par de datos, la mediana ser√° la **media de los dos del centro**.  
Ejemplo con 12 datos:

```

77, 78, 85, 86, 86, 86, 87, 87, 94, 98, 99, 103

```

Los dos del medio son 86 y 87 ‚Üí $(86+87)/2 = 86.5$.

---

### üîπ Moda (Mode)
La Moda es el valor que aparece con mayor frecuencia en el conjunto de datos

En nuestro caso:

```

99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86

```

El valor **86** aparece **3 veces**, as√≠ que la moda es **86**.

---

Ejemplo en Python con sueldos


`````python
import numpy as np
from scipy import stats

# Datos: la mayor√≠a sueldos normales y uno muy alto
sueldos = [1200, 1300, 1250, 1400, 1350, 1280, 1500, 1250, 1300, 30000]

# Medidas
media = np.mean(sueldos)
mediana = np.median(sueldos)
moda = stats.mode(sueldos, keepdims=True)[0][0]

print("Media:", media)
print("Mediana:", mediana)
print("Moda:", moda)

`````

üìå **Conclusi√≥n**

* La **media** nos da una idea del valor promedio.
* La **mediana** nos dice cu√°l es el valor central (muy √∫til si hay valores extremos que distorsionan la media).
* La **moda** muestra el valor m√°s frecuente.

El uso de estas medidas es indispensable para **entender los datos antes de entrenar un modelo**, lo que se traduce en t√©cnicas de preprocesamiento esenciales para la fase inicial del ML:

*   **Limpieza de Datos:** Antes de que un algoritmo pueda aprender, debe comprender las propiedades estad√≠sticas de los datos, incluyendo su distribuci√≥n y varianza. Un buen manejo de la estad√≠stica permite tareas de preprocesamiento como la limpieza de datos y la remoci√≥n de valores at√≠picos.
*   **Normalizaci√≥n y Estandarizaci√≥n:** Estas medidas se utilizan para aplicar **t√©cnicas de normalizaci√≥n** (como la estandarizaci√≥n por puntuaci√≥n z), que son herramientas matem√°ticas que ajustan los datos para que tengan una media de cero y una desviaci√≥n est√°ndar de uno. Este proceso asegura que todas las caracter√≠sticas est√©n en una escala similar, impidiendo que una caracter√≠stica, debido a su magnitud, **domine el proceso de aprendizaje**.
*   **Funciones de P√©rdida (Loss Functions):** Estos conceptos descriptivos se integran directamente en las funciones de optimizaci√≥n. Por ejemplo, el **Error Cuadr√°tico Medio (MSE)**, una funci√≥n de p√©rdida com√∫n en regresi√≥n, se define en t√©rminos del **valor esperado o media** y cuantifica la varianza de los errores de predicci√≥n.

En esencia, la estad√≠stica descriptiva sienta las bases matem√°ticas para que los algoritmos de **C√°lculo Multivariable** (como el descenso de gradiente) y **√Ålgebra Lineal** puedan operar de manera eficiente y precisa en la b√∫squeda de patrones y la optimizaci√≥n de las redes neuronales.

# 3.1 Medidas de Dispersi√≥n

Hasta ahora hemos visto medidas que nos dicen **d√≥nde est√°n los datos** (media, mediana, moda).

## üîπ Rango

El rango, tambi√©n llamado span, es la distancia entre el m√≠nimo y el m√°ximo de una distribuci√≥n, es decir, la distancia entre el valor m√°s peque√±o y el m√°s grande.


![](https://numiqo.es/assets/tutorial/Rango.png)


### Ejemplo

`````python
import numpy as np

notas = [5, 7, 8, 6, 7, 10, 4, 7, 9, 6]
rango = np.max(notas) - np.min(notas)

print("Rango:", rango)
`````

üìå El rango muestra la amplitud total de los datos.


## üîπ Desviaci√≥n T√≠pica ($\sigma$)

La **desviaci√≥n t√≠pica (o est√°ndar)** es la **ra√≠z cuadrada de la varianza** . Esta medida es preferida para la interpretaci√≥n pr√°ctica porque est√° expresada en las **mismas unidades** que los datos originales .

*   Una **desviaci√≥n baja** indica que los datos est√°n agrupados muy **cerca de la media** .
*   Una **desviaci√≥n alta** indica que los datos est√°n **muy dispersos** o separados entre s√≠.

![](https://datatab.es/assets/tutorial/Desviaci%C3%B3n_est%C3%A1ndar.png)
---

- Desviaci√≥n t√≠pica = **0.9**

## üîπ Varianza ($\text{Var}[X]$ o $\sigma^2$)

La **varianza** mide la desviaci√≥n cuadr√°tica promedio de los datos respecto a la media. En el contexto probabil√≠stico, la varianza es el **valor esperado de la desviaci√≥n al cuadrado** de una variable aleatoria ($X$) respecto a su media ($\mu$):

$$\text{Var}[X] = \mathbb{E}[(X - \mu)^2]$$

La f√≥rmula de la varianza para un conjunto de datos finito ($n$ observaciones) es:

$$\sigma^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2$$

El prop√≥sito de elevar al cuadrado las diferencias es eliminar los signos negativos y penalizar las desviaciones grandes con mayor peso, una t√©cnica que se traslada directamente a las funciones de p√©rdida en ML (como se ver√° a continuaci√≥n).

Como los valores se elevan al cuadrado, el resultado tiene una unidad diferente (la unidad al cuadrado) que los valores originales. Por lo tanto, es dif√≠cil relacionar los resultados.

---

## Varianza frente a desviaci√≥n est√°ndar

As√≠ pues, la diferencia entre la varianza y la desviaci√≥n est√°ndar del par√°metro de dispersi√≥n es que la desviaci√≥n est√°ndar mide la distancia media respecto a la media y la varianza mide la distancia media al cuadrado respecto a la media. En otras palabras, la varianza es la desviaci√≥n est√°ndar elevada al cuadrado y la desviaci√≥n est√°ndar es la ra√≠z de la varianza.

Sin embargo, esta elevaci√≥n al cuadrado da lugar a un ratio dif√≠cil de interpretar, ya que la unidad no se corresponde con los datos originales. Por este motivo, es aconsejable utilizar siempre la desviaci√≥n t√≠pica para describir una muestra, ya que facilita la interpretaci√≥n.

---


## 3.2. El Impacto de la Dispersi√≥n en Machine Learning (ML)

En **Machine Learning**, la desviaci√≥n t√≠pica y la varianza son indispensables para la fase de preprocesamiento de datos y para la etapa de optimizaci√≥n del modelo [Query]. La estad√≠stica es una de las primeras herramientas matem√°ticas utilizadas en el procesamiento de datos de IA.

### A. Funci√≥n de P√©rdida y Optimizaci√≥n

Las medidas de dispersi√≥n se encuentran en el coraz√≥n de la optimizaci√≥n de los modelos de IA, que es el proceso de ajustar los par√°metros (pesos) para minimizar una funci√≥n de error.

*   **Error Cuadr√°tico Medio (MSE):** La varianza est√° directamente relacionada con una de las funciones de p√©rdida m√°s comunes en los modelos de regresi√≥n, el **Error Cuadr√°tico Medio (MSE)**. El MSE esencialmente mide la **varianza de los errores** de predicci√≥n. Un MSE bajo (es decir, una varianza baja del error) indica que las predicciones del modelo son mejores.
*   **Cuantificaci√≥n de la Incertidumbre:** La varianza y la desviaci√≥n t√≠pica son vitales para **evaluar el rendimiento del modelo** y **cuantificar la incertidumbre** en las predicciones de la IA. Una baja varianza en las predicciones sugiere una alta confianza del modelo, mientras que una alta varianza indica incertidumbre.

### B. Normalizaci√≥n y Estandarizaci√≥n de Datos

La dispersi√≥n es clave en el proceso de **preprocesamiento** de datos, una etapa crucial **antes de entrenar un modelo**:

*   **Estandarizaci√≥n (Normalizaci√≥n Z-score):** Para que un modelo de ML (especialmente las redes neuronales) aprenda de manera efectiva, es com√∫n estandarizar los datos para que tengan una **media de cero y una desviaci√≥n est√°ndar de uno** ($\sigma=1$).
*   **Prevenci√≥n de Dominio de Caracter√≠sticas:** Este proceso matem√°tico (que es una transformaci√≥n lineal) asegura que todas las caracter√≠sticas est√©n en una escala similar, **impidiendo que una sola caracter√≠stica domine el proceso de aprendizaje** debido a su magnitud (por ejemplo, si el "salario" tiene valores de miles y la "edad" valores de decenas).

![](https://www.mygreatlearning.com/blog/wp-content/uploads/2025/03/salary-before-standardization.jpg.webp)

![](https://www.mygreatlearning.com/blog/wp-content/uploads/2025/03/data-after-standardization.jpg.webp)


### C. Detecci√≥n de Valores At√≠picos (*Outliers*)

La Desviaci√≥n T√≠pica sirve como una m√©trica robusta para identificar si una variable tiene valores "normales" o si presenta **outliers**.

*   Los **valores at√≠picos** son puntos de datos que se desv√≠an significativamente del resto.
*   T√≠picamente, cualquier punto que se encuentre a m√°s de **2 desviaciones est√°ndar** ($\pm 2\sigma$) de la media se considera un valor at√≠pico.

![](https://statistics.laerd.com/spss-tutorials/img/lr/outliers.png)

### D. Reducci√≥n de Dimensionalidad (PCA)

La varianza es el concepto central de t√©cnicas avanzadas como el **An√°lisis de Componentes Principales (PCA)**.

*   PCA, que es crucial para manejar datos de alta dimensi√≥n (como im√°genes o texto), primero requiere que los datos sean estandarizados (usando media y desviaci√≥n t√≠pica).
*   PCA busca las direcciones (autovectores) en el espacio de datos donde la **varianza es m√°xima**. Los autovalores correspondientes indican la magnitud de esa varianza. Al seleccionar solo los autovectores con las mayores varianzas (los autovalores m√°s grandes), PCA reduce la dimensionalidad mientras conserva la informaci√≥n m√°s significativa.

Los cuartiles dividen los datos en cuatro partes, lo m√°s iguales posible. Para el c√°lculo de los cuartiles, los datos deben estar ordenados del valor m√°s peque√±o al m√°s grande.

- Cuartil (Q2): La mediana de los datos, es decir, el 50% de los valores son menores y el 50% de los valores son mayores.
- Cuartil (Q3): El valor medio entre la mediana y el valor m√°s grande (m√°ximo).

## üîπ Percentiles

Los percentiles dividen un conjunto de datos ordenado en 100 partes iguales. Estas divisiones permiten entender la posici√≥n relativa de un dato dentro del conjunto.

## üîπ Rango intercuart√≠lico (IQR)
A diferencia del rango en el que se encuentra el 100% de todos los valores, a menudo se desea conocer el rango en el que se encuentra el 50% medio de todos los valores. Este par√°metro de dispersi√≥n se denomina rango intercuartil. Por lo tanto, el 25% superior e inferior de los valores no se tienen en cuenta para el rango intercuart√≠lico.

![](https://numiqo.es/assets/tutorial/Rango-intercuart%C3%ADlico.png)

El **IQR** es muy √∫til para detectar valores at√≠picos (outliers). Cualquier valor menor al l√≠mite inferior o mayor al l√≠mite superior se marca como outlier.


![](https://images.squarespace-cdn.com/content/v1/5b872f96aa49a1a1da364999/f421dc04-fbd0-4ac3-bb4a-15f2c20fd072/iqr_llustrated.png)

`````python
notas = [5, 7, 8, 6, 7, 10, 4, 7, 9, 6]

q1 = np.percentile(notas, 25)
q2 = np.percentile(notas, 50)  # mediana
q3 = np.percentile(notas, 75)
iqr = q3 - q1

print("Q1:", q1, "Q2:", q2, "Q3:", q3)
print("IQR:", iqr)
`````

## üîπ Histogramas

Un **histograma** es un gr√°fico que nos permite ver c√≥mo se distribuyen los datos en intervalos (llamados **bins**).  

### 1. V√≠nculo con la Probabilidad y el C√°lculo

El concepto fundamental que subyace al histograma es la **funci√≥n de densidad de probabilidad (PDF)**, un elemento central de la probabilidad y la estad√≠stica.

*   **Bins y PDF:** Los histogramas dividen el rango de valores de una variable en **intervalos (bins)** . La altura de cada barra de frecuencia es una aproximaci√≥n visual de la funci√≥n de densidad de probabilidad. Para las variables continuas, la probabilidad se encuentra integrando la PDF sobre un rango de valores. Un histograma bien construido intenta emular la forma de esta funci√≥n matem√°tica subyacente.
*   **Distribuciones Notables:** Los histogramas permiten visualizar si la distribuci√≥n de una caracter√≠stica se ajusta a distribuciones te√≥ricas comunes, como la **Distribuci√≥n Normal (Gaussiana)** o "Campana de Gauss", crucial para muchos modelos de ML que asumen un comportamiento normal.

### 2. Detecci√≥n de Patrones y Anomal√≠as (Outliers)

La detecci√≥n de **valores at√≠picos (*outliers*)** es cr√≠tica en el preprocesamiento de datos, ya que estos pueden distorsionar significativamente la **Media** y la **Varianza**, afectando la precisi√≥n del modelo.

*   **Concentraciones:** Las concentraciones m√°s altas en el histograma reflejan las **medidas de tendencia central** (Moda y Media) [Query].
*   **Detecci√≥n Visual de Outliers:** Los *outliers* se manifiestan visualmente en el histograma como barras de muy baja frecuencia ubicadas lejos de la concentraci√≥n principal de los datos. Esta detecci√≥n visual apoya el an√°lisis estad√≠stico m√°s riguroso, que define un *outlier* t√≠picamente como un punto a m√°s de **2 desviaciones est√°ndar ($\sigma$)** de la media.
*   **Asimetr√≠a (*Skewness*):** Los histogramas revelan si la distribuci√≥n es sim√©trica (como la Gaussiana) o asim√©trica. Si una distribuci√≥n est√° sesgada, la media es un descriptor menos fiable que la mediana.

En **Machine Learning**, estas medidas se usan mucho para **entender los datos antes de entrenar un modelo**. La visualizaci√≥n mediante histogramas es un paso previo indispensable para la optimizaci√≥n de los modelos:

*   **Estandarizaci√≥n y Normalizaci√≥n:** La mayor√≠a de los algoritmos de optimizaci√≥n (como el **Descenso de Gradiente**) funcionan mejor si los datos est√°n estandarizados (centrados en una media de cero y con una desviaci√≥n est√°ndar de uno). Los histogramas permiten verificar visualmente que las caracter√≠sticas est√°n en una **escala similar**, evitando que una caracter√≠stica "domine el proceso de aprendizaje debido a su magnitud".
*   **An√°lisis de Caracter√≠sticas (Feature Analysis):** En aplicaciones que manejan datos de alta dimensi√≥n (como el *deep learning*) es crucial analizar las caracter√≠sticas individuales antes de aplicar t√©cnicas de reducci√≥n de dimensionalidad como **PCA** (An√°lisis de Componentes Principales), que se basa en la **m√°xima varianza** en los datos. # Boxplot (Diagrama de caja)

El **Diagrama de Caja (Boxplot)** es una herramienta gr√°fica de la estad√≠stica descriptiva que resulta inestimable en la fase de **An√°lisis Exploratorio de Datos (EDA)**, un paso previo e indispensable **antes de entrenar un modelo de Machine Learning** .

Este gr√°fico es fundamental porque ofrece, en una sola visualizaci√≥n, una perspectiva completa sobre la **dispersi√≥n**, la **simetr√≠a** y, lo m√°s cr√≠tico para los modelos de IA, la presencia de **valores at√≠picos (*outliers*)**

A continuaci√≥n, se detalla la composici√≥n y la relevancia matem√°tica y de IA del *boxplot*:

### 1. Componentes del Diagrama de Caja

El diagrama de caja se construye a partir de los cuantiles (percentiles) que dividen el conjunto de datos ordenado en cuatro partes iguales, y proporciona una medida de dispersi√≥n que es **robusta** ante valores extremos:

![](https://miro.medium.com/v2/resize:fit:1200/1*0MPDTLn8KoLApoFvI0P2vQ.png)

### 2. Ventaja en el An√°lisis de Datos (EDA)

La principal ventaja del *boxplot* es que **permite ver dispersi√≥n, simetr√≠a y valores at√≠picos en un solo gr√°fico** .

*   **Dispersi√≥n y Variabilidad:** La longitud de la caja (IQR) y la extensi√≥n de los bigotes ofrecen una visualizaci√≥n inmediata de cu√°n agrupados o dispersos est√°n los datos.
*   **Simetr√≠a:** Si la mediana est√° en el centro de la caja y los bigotes son de longitud similar, la distribuci√≥n tiende a ser sim√©trica (aproxim√°ndose a una distribuci√≥n normal o Gaussiana).

### 3. El Boxplot y la Detecci√≥n de *Outliers* en Machine Learning

La capacidad del diagrama de caja para identificar valores at√≠picos es crucial en ML, ya que estos puntos de datos pueden impactar dr√°sticamente el proceso de entrenamiento.

1.  **Distorsi√≥n de la Media y la Varianza:** Los valores at√≠picos son datos que se desv√≠an significativamente. Si no se manejan, los *outliers* pueden afectar gravemente la **media ($\mu$)** y la **Varianza ($\sigma^2$)**, lo cual es problem√°tico porque la optimizaci√≥n en las redes neuronales a menudo se basa en minimizar errores cuadr√°ticos (como el MSE), que dependen de la varianza.
2.  **M√©trica de Detecci√≥n:** El IQR proporciona un m√©todo *robusto* para establecer umbrales de atipicidad. Sin embargo, en la pr√°ctica de IA, la detecci√≥n de *outliers* se relaciona com√∫nmente con la **Desviaci√≥n T√≠pica ($\sigma$)**, otra medida de dispersi√≥n:

    *   T√≠picamente, cualquier punto de dato que se encuentre a **m√°s de 2 desviaciones est√°ndar ($\sigma$) de la media ($\mu$) se considera un *outlier***.

3.  **Preprocesamiento y Estandarizaci√≥n:** La estad√≠stica descriptiva (media y desviaci√≥n t√≠pica) se utiliza para la **estandarizaci√≥n** o **normalizaci√≥n** de los datos, ajust√°ndolos para que tengan una media de cero y una desviaci√≥n est√°ndar de uno. Este paso es vital para asegurar que todas las caracter√≠sticas tengan la misma influencia en el modelo y que ninguna caracter√≠stica "domine el proceso de aprendizaje". La visualizaci√≥n con *boxplots* puede guiar al analista sobre si los datos deben ser estandarizados o si los *outliers* deben ser eliminados antes de aplicar esta transformaci√≥n.

`````python
import matplotlib.pyplot as plt

notas = [4,5,6,7,8,5,6,7,9,10,6,7,5,4,8,9,10,6,7,8,0,20]

plt.boxplot(notas)
plt.title("Boxplot de notas de alumnos con outliers")
plt.show()
`````

üìå Interpretaci√≥n:

* La **l√≠nea central** indica la mediana.
* La **caja** indica el rango intercuart√≠lico (la mayor√≠a de las notas).
* Los **bigotes** se extienden a los valores m√≠nimo y m√°ximo que no son at√≠picos.
* Los outliners se marcan como puntos fuera de la caja (0 y 20).

# 5 Correlaci√≥n y Covarianza

Cuando trabajamos con **dos variables**, no solo nos interesa su distribuci√≥n individual,  
sino tambi√©n **c√≥mo se relacionan entre s√≠**.

---

## üîπ Covarianza

- La **covarianza** mide c√≥mo var√≠an dos variables juntas.  
- Si aumenta una y tambi√©n lo hace la otra ‚Üí covarianza positiva.  
- Si una aumenta y la otra disminuye ‚Üí covarianza negativa.  
- Si no hay patr√≥n ‚Üí covarianza ‚âà 0.  

üìå Problema: depende de las unidades, por lo que no es f√°cil comparar.

---

## üîπ Correlaci√≥n

- La **correlaci√≥n** es una medida estandarizada de la relaci√≥n entre dos variables.  
- Valores entre **-1 y 1**:  
  - +1 ‚Üí correlaci√≥n positiva perfecta.  
  - -1 ‚Üí correlaci√≥n negativa perfecta.  
  - 0 ‚Üí sin relaci√≥n.  

üìå La correlaci√≥n es m√°s √∫til porque no depende de las unidades de medida.

---

## üîπ Ejemplo: Tama√±o de vivienda vs Precio

Queremos analizar si las viviendas m√°s grandes suelen ser m√°s caras.

`````python
import numpy as np
import pandas as pd

# Datos ficticios: tama√±o en m¬≤ y precio en miles de ‚Ç¨
tamanio = [50, 60, 80, 100, 120, 150, 200]
precio  = [100,120,150,180,210,250,320]

# Covarianza con NumPy
cov = np.cov(tamanio, precio, ddof=0)
print("Matriz de covarianza:\n", cov)

# Correlaci√≥n con NumPy
corr = np.corrcoef(tamanio, precio)
print("\nMatriz de correlaci√≥n:\n", corr)

# Con Pandas
df = pd.DataFrame({"Tama√±o": tamanio, "Precio": precio})
print("\nResumen con Pandas:")
print(df.corr())
`````

---

## üîé Explicaci√≥n de resultados

* La **covarianza** ser√° positiva ‚Üí cuando aumenta el tama√±o, tambi√©n aumenta el precio.
* La **correlaci√≥n** estar√° cerca de **+1** ‚Üí relaci√≥n fuerte y positiva.

üìå En un scatter plot ver√≠amos los puntos alineados en forma ascendente.

`````python

import matplotlib.pyplot as plt

plt.scatter(tamanio, precio)
plt.title("Tama√±o de vivienda vs Precio")
plt.xlabel("Tama√±o (m¬≤)")
plt.ylabel("Precio (miles de ‚Ç¨)")
plt.show()
`````

‚úÖ **Idea clave:**

* La **covarianza** nos dice si dos variables suben o bajan juntas.
* La **correlaci√≥n** nos dice adem√°s **qu√© tan fuerte es esa relaci√≥n**.
* En Machine Learning, estas medidas nos ayudan a:

  * Seleccionar variables relevantes (ej. si dos variables est√°n muy correlacionadas, puede bastar con una).
  * Entender qu√© factores influyen en un resultado.
